```python
import os
from dotenv import load_dotenv

load_dotenv()  # load all the environment variables
```

```python
# Read API key from .env file in current working directory
# We shouldn't clear the API key first since load_dotenv() was already called above
# Just verify the API key is set correctly
if not os.getenv("OPENAI_API_KEY"):
    env_path = os.path.join(os.getcwd(), ".env")
    if os.path.exists(env_path):
        load_dotenv(env_path)
        if not os.getenv("OPENAI_API_KEY"):
            print("OPENAI_API_KEY not found in .env file")
    else:
        print("No .env file found in current directory")

print("API Key loaded:", bool(os.getenv("OPENAI_API_KEY")))
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(
    model="text-embedding-3-large", openai_api_key=os.getenv("OPENAI_API_KEY")
)
embeddings
```

```python
text = "This is a tutorial on OPENAI embedding"
query_result = embeddings.embed_query(text)
query_result
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("speech.txt")
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_documents = text_splitter.split_documents(docs)
final_documents
```

```python
## Vector Embedding And Vector StoreDB
## chroma is a vector store that is used to store the vectors and documents in a database
# Import Chroma vector store from LangChain community

from langchain_community.vectorstores import Chroma

# Create a Chroma vector store from our documents using the 1024-dimensional embeddings
# This will:
# 1. Convert each document into vector embeddings using embeddings_1024
# 2. Store the vectors and documents in a Chroma database
# 3. Enable similarity search on the documents
db = Chroma.from_documents(final_documents, embeddings_1024)
db
```

```python
### Retrieve the results from query vectorstore db
query = "It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results = db.similarity_search(query)
retrieved_results
```
