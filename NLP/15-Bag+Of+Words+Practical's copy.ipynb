```python
import pandas as pd
messages=pd.read_csv('sms_spam.csv')
messages.columns = ["label","message"]
```

```python
messages
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Create Bag Of Words
        -- there are 2 types of BOG: normal and binary
        -- in binary, its just 0 and 1
        -- in normal: you can keep the number as the frequency of that word

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer

## for Binary BOW enable binary=True 
cv=CountVectorizer(max_features=100,binary=True)
## just take the 100 most occuring words
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
X
```

### N-Grams

```python
cv.vocabulary_
```

```python
## Create the Bag OF Words model with ngram
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=100,binary=True,ngram_range=(2,3))
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```
